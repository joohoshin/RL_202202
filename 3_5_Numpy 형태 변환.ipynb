딥러닝 모델링 시 입력 데이터의 형태를 변환해줘야 하는 경우가 많으며
이에 활용되는 함수를 알아보자

```python
import numpy as np
```

```python
# 이미지 데이터의 경우 대부분 4개의 차원을 가짐
# 텐서플로우는 (n, w, h, c) 형태로 모델링
```

```python
arr1 = np.random.randint(0,255, (30,30,3))
```

```python
arr1.shape
```

```python
arr1
```

expand_dims: 차원을 추가할 수 있다

```python
# 1차원 추가하기
arr2 = np.expand_dims(arr1,0)  # 몇 번째에 추가할지 지정 가능
```

```python
?np.expand_dims
```

```python
arr2.shape
```

```python
arr3 = np.expand_dims(arr1,3)  # 몇 번째에 추가할지 지정 가능
arr3.shape
```

squeeze: 차원 삭제

```python
# 차원 삭제하기
arr3.squeeze(3).shape
```

```python
np.arange(10)
```

reshape: 형태 변경

```python
arr4 = np.arange(100).reshape(20,5)
```

```python
arr4.shape
```

```python
arr4
```
